```python
#  Bar Plots for Categorical Variables

import seaborn as sns
import matplotlib.pyplot as plt

# Set visual theme
sns.set(style="whitegrid")

# Identify categorical columns
categorical_columns = [col for col in df.columns if df[col].dtype == 'object']

# Create bar plots for each categorical column
for col in categorical_columns:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, data=df)
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
#  Histograms for Numerical Variables

# Select only numeric columns
numerical_columns = df.select_dtypes(include='number').columns

# Plot histograms for each numeric column
for col in numerical_columns:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col].dropna(), kde=True)  # Drop NA to avoid errors
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.tight_layout()
    plt.show()
```

```python
# Correlation Heatmap

# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include='number')

# Calculate correlation matrix
correlation = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("🔗 Correlation Heatmap of Numerical Features")
plt.tight_layout()
plt.show()
```

```python
# Display rows where Age is still missing
missing_rows = df[df['Age'].isnull()]
print(missing_rows[['Pclass', 'SibSp', 'Parch', 'Sex', 'Fare', 'Name']])
```

```python
df['Age'] = df['Age'].fillna(df.groupby(['Pclass', 'Sex'])['Age'].transform('mean'))
```
